# 3. DATA CLEANING

In this section we are going to prepare data for analysis by:
- Filtering countries of interest
- Droping rows with missing values
- Convert date column to datetime: pd.to_datetime().
- Handle missing numeric values with fillna() or interpolate().

```python
# filter countries of interest
countries_of_interest = ['United States','India','Kenya']
# add countries of interest as a  column
covid_df['countries_of_interest'] = covid_df['location'].apply(lambda x: x if x in countries_of_interest else None)
```

```python
# drop rows with missing values
covid_df = covid_df.dropna(subset=['total_cases', 'total_deaths', 'total_vaccinations'])
```

```python
# convert date column to datetime
covid_df['date'] = pd.to_datetime(covid_df['date'])
```

```python
#handle missing values
covid_df['total_cases'] = covid_df['total_cases'].fillna(0)
```

# 4. EXPLORATORY DATA ANALYSIS (EDA)

In this section we are going to generate descriptive statistics and explore trends by:
- Plotting total cases over time for selected countries
- total deaths over time
- comparing daily new cases between countries
- calculating the death_rate

```python
# linechart cases vs deaths over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=covid_df, x='date', y='total_cases', hue='countries_of_interest')
plt.title('Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.show()

```

```python
#deaths over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=covid_df, x='date', y='total_deaths', hue='countries_of_interest')
plt.title('Total Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
# bar graph for daily new cases vs countries
top_20_countries = covid_df.groupby('location')['new_cases'].sum().nlargest(20).index
top_20_df = covid_df[covid_df['location'].isin(top_20_countries)]
plt.figure(figsize=(12, 6))
sns.barplot(data=top_20_df, x='location', y='new_cases', estimator=sum)
plt.title('Top 20 Countries with Highest Daily New Cases')
plt.xlabel('Country')
plt.ylabel('Daily New Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# top countries with most cases bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=top_20_df, x='location', y='total_cases', palette='viridis')
plt.title('Top 10 Countries with Most Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# calculate the death rate
covid_df['death_rate'] = (covid_df['total_deaths'] / covid_df['total_cases']) * 100
covid_df['death_rate']
```

# VISUALIZING VACCINATION PROGRESS

The goal of this section is to analyze the vaccination roll outs

## TASKS
- Plot cumulative vaccinations over time for selected countries.
- Compare % vaccinated population


```python
# LINE CHART FOR VACCINATION OVER TIME FOR SELECTED COUNTRIES
plt.figure(figsize=(12, 6))
sns.lineplot(data=covid_df, x='date', y='total_vaccinations', hue='location')
plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
# vaccination per location
plt.figure(figsize=(12, 6))
sns.barplot(data=top_20_df, x='location', y='total_vaccinations', palette='viridis')
plt.title('Total Vaccinations by Country')
plt.xlabel('Country')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# PIE CHART FOR VACCINATED VS UNVACCINATED
vaccinated = covid_df[covid_df['total_vaccinations'] > 0].shape[0]
unvaccinated = covid_df[covid_df['total_vaccinations'] == 0].shape[0]
labels = ['Vaccinated', 'Unvaccinated']
sizes = [vaccinated, unvaccinated]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66c2a5', '#fc8d62'])
plt.title('Vaccination Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

# Insights and Findings

## Total Cases Over Time (Selected Countries)*
Insight: Countries like the United States and India show a steep and continuous rise in total cases, reflecting large populations and widespread outbreaks.

Finding: Kenya shows a much slower increase, which may be due to underreporting, fewer tests, or lower infection spread.
The shape of the curve shows how quickly the virus spread in each country.

## Total Deaths Over Time
Insight: United States and India also show high cumulative deaths, mirroring their total case counts.

Finding: Death curves usually lag behind case curves, especially after a large wave.


## Daily New Cases Comparison
Insight: New daily cases spiked sharply in India during mid-2021 (Delta variant) and again in early 2022 (Omicron).

Finding: USA had multiple waves, but vaccination helped reduce the height and length of subsequent waves.

Kenya’s daily cases remained lower overall but show visible surges corresponding to global waves.


## Death Rate (total_deaths / total_cases)
Insight: Some countries like Mexico and South Africa have high case fatality rates (CFR) — possibly due to limited healthcare access or underreporting of mild cases.

Finding: Countries with strong healthcare systems and widespread testing, like Germany or South Korea, typically show lower death rates.

CFR can be misleading if testing is low (deaths stay accurate, cases undercounted).

## Vaccination Progress 
Insight: High-income countries like the USA, UK, and Germany vaccinated large portions of their populations early.

Finding: Lower-income countries like Kenya lagged behind initially but gradually caught up with partial or full vaccination coverage.

There’s often a correlation between higher vaccination and lower death rates post-2021.
